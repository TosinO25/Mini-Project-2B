# Emotion Recognition - Results Analysis and Visualization

Load results from training and create plots for your report.

## Learning Objectives
- Analyze scaling trends in neural networks
- Understand accuracy vs efficiency trade-offs
- Visualize performance metrics
- Interpret results in context of scaling laws

## Part 1: Load Results

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load your results CSV
results = pd.read_csv('emotion_model_results.csv')
print("Results loaded:")
print(results.to_string())
```

## Part 2: Inspect Results Table

```python
# Display summary statistics
print("\nResults Summary:")
print(f"Number of models: {len(results)}")
print(f"\nAccuracy Range:")
print(f"  Train: {results['train_accuracy'].min():.4f} - {results['train_accuracy'].max():.4f}")
print(f"  Test: {results['test_accuracy'].min():.4f} - {results['test_accuracy'].max():.4f}")
print(f"\nTraining Time Range:")
print(f"  Min: {results['train_time_seconds'].min():.2f}s")
print(f"  Max: {results['train_time_seconds'].max():.2f}s")
print(f"  Ratio (max/min): {results['train_time_seconds'].max() / results['train_time_seconds'].min():.1f}x")
print(f"\nParameters Range:")
print(f"  Min: {results['num_parameters'].min():,}")
print(f"  Max: {results['num_parameters'].max():,}")
```

## Part 3: Plot 1 - Accuracy vs Model Size

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Plot both train and test accuracy
ax.plot(range(len(results)), results['train_accuracy'], 'o-', label='Train Accuracy', linewidth=2, markersize=8)
ax.plot(range(len(results)), results['test_accuracy'], 's-', label='Test Accuracy', linewidth=2, markersize=8)

# Formatting
ax.set_xlabel('Model (increasing size)', fontsize=12, fontweight='bold')
ax.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax.set_title('Accuracy vs Model Size', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(results)))
ax.set_xticklabels(results['model_name'], rotation=45)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_ylim([0, 1.05])

# Add value labels on points
for i, (train, test) in enumerate(zip(results['train_accuracy'], results['test_accuracy'])):
    ax.text(i, train + 0.02, f'{train:.3f}', ha='center', fontsize=9)
    ax.text(i, test - 0.05, f'{test:.3f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('plot1_accuracy_vs_size.png', dpi=300, bbox_inches='tight')
plt.show()

print("Plot 1 saved: plot1_accuracy_vs_size.png")
```

## Part 4: Plot 2 - Training Time vs Model Size

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for training time
colors = plt.cm.viridis(np.linspace(0, 1, len(results)))
bars = ax.bar(range(len(results)), results['train_time_seconds'], color=colors, alpha=0.8, edgecolor='black')

# Formatting
ax.set_xlabel('Model (increasing size)', fontsize=12, fontweight='bold')
ax.set_ylabel('Training Time (seconds)', fontsize=12, fontweight='bold')
ax.set_title('Training Time vs Model Size', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(results)))
ax.set_xticklabels(results['model_name'], rotation=45)
ax.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for i, (bar, time_val) in enumerate(zip(bars, results['train_time_seconds'])):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{time_val:.1f}s',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('plot2_training_time.png', dpi=300, bbox_inches='tight')
plt.show()

print("Plot 2 saved: plot2_training_time.png")
```

## Part 5: Plot 3 - Test Time vs Model Size

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for test time
colors = plt.cm.plasma(np.linspace(0, 1, len(results)))
bars = ax.bar(range(len(results)), results['test_time_seconds'], color=colors, alpha=0.8, edgecolor='black')

# Formatting
ax.set_xlabel('Model (increasing size)', fontsize=12, fontweight='bold')
ax.set_ylabel('Test Time (seconds)', fontsize=12, fontweight='bold')
ax.set_title('Test Time vs Model Size', fontsize=14, fontweight='bold')
ax.set_xticks(range(len(results)))
ax.set_xticklabels(results['model_name'], rotation=45)
ax.grid(True, alpha=0.3, axis='y')

# Add value labels
for i, (bar, time_val) in enumerate(zip(bars, results['test_time_seconds'])):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{time_val:.2f}s',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('plot3_test_time.png', dpi=300, bbox_inches='tight')
plt.show()

print("Plot 3 saved: plot3_test_time.png")
```

## Part 6: Plot 4 - Accuracy vs Number of Parameters

```python
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot: parameters vs accuracy
ax.scatter(results['num_parameters'], results['train_accuracy'], 
           s=200, alpha=0.6, label='Train Accuracy', marker='o')
ax.scatter(results['num_parameters'], results['test_accuracy'], 
           s=200, alpha=0.6, label='Test Accuracy', marker='s')

# Add trend lines
z_train = np.polyfit(results['num_parameters'], results['train_accuracy'], 2)
p_train = np.poly1d(z_train)
x_smooth = np.linspace(results['num_parameters'].min(), results['num_parameters'].max(), 100)
ax.plot(x_smooth, p_train(x_smooth), '--', alpha=0.5, color='blue', linewidth=2)

z_test = np.polyfit(results['num_parameters'], results['test_accuracy'], 2)
p_test = np.poly1d(z_test)
ax.plot(x_smooth, p_test(x_smooth), '--', alpha=0.5, color='orange', linewidth=2)

# Formatting
ax.set_xlabel('Number of Parameters', fontsize=12, fontweight='bold')
ax.set_ylabel('Accuracy', fontsize=12, fontweight='bold')
ax.set_title('Accuracy vs Model Parameters', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)
ax.set_ylim([0, 1.05])

# Format x-axis for large numbers
ax.ticklabel_format(style='plain', axis='x')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{int(x/1000)}K' if x >= 1000 else f'{int(x)}'))

plt.tight_layout()
plt.savefig('plot4_accuracy_vs_parameters.png', dpi=300, bbox_inches='tight')
plt.show()

print("Plot 4 saved: plot4_accuracy_vs_parameters.png")
```

## Part 7: Analysis and Insights

```python
print("ANALYSIS INSIGHTS")
print("="*70)

# Find best models
best_test_idx = results['test_accuracy'].idxmax()
best_train_idx = results['train_accuracy'].idxmax()
fastest_idx = results['train_time_seconds'].idxmin()

print(f"\n1. ACCURACY TRENDS:")
print(f"   Best test accuracy: {results.loc[best_test_idx, 'test_accuracy']:.4f} ({results.loc[best_test_idx, 'model_name']})")
print(f"   Best train accuracy: {results.loc[best_train_idx, 'train_accuracy']:.4f} ({results.loc[best_train_idx, 'model_name']})")
print(f"   Accuracy gap (train-test) at best model: {results.loc[best_test_idx, 'train_accuracy'] - results.loc[best_test_idx, 'test_accuracy']:.4f}")

# Check for overfitting trend
print(f"\n2. OVERFITTING ANALYSIS:")
max_gap = (results['train_accuracy'] - results['test_accuracy']).max()
max_gap_idx = (results['train_accuracy'] - results['test_accuracy']).idxmax()
print(f"   Maximum train-test gap: {max_gap:.4f} ({results.loc[max_gap_idx, 'model_name']})")
print(f"   Indicates: {'Strong overfitting' if max_gap > 0.1 else 'Moderate overfitting' if max_gap > 0.05 else 'Minimal overfitting'}")

# Runtime trends
print(f"\n3. COMPUTATIONAL EFFICIENCY:")
print(f"   Fastest model: {results.loc[fastest_idx, 'model_name']} ({results.loc[fastest_idx, 'train_time_seconds']:.2f}s)")
print(f"   Slowest model: {results.loc[results['train_time_seconds'].idxmax(), 'model_name']} ({results['train_time_seconds'].max():.2f}s)")
time_increase = results['train_time_seconds'].max() / results['train_time_seconds'].min()
print(f"   Training time increase: {time_increase:.1f}x")

# Scaling law check
print(f"\n4. SCALING LAW OBSERVATIONS:")
param_increase = results['num_parameters'].max() / results['num_parameters'].min()
acc_increase = results['test_accuracy'].max() / results['test_accuracy'].min()
print(f"   Parameter increase: {param_increase:.1f}x")
print(f"   Accuracy increase: {acc_increase:.2f}x")
print(f"   Time-to-accuracy ratio: {time_increase / acc_increase:.2f}x")

# Recommendation
print(f"\n5. RECOMMENDATION:")
# Find best accuracy per second ratio
results['acc_per_second'] = results['test_accuracy'] / results['train_time_seconds']
best_ratio_idx = results['acc_per_second'].idxmax()
print(f"   Best accuracy-per-second: {results.loc[best_ratio_idx, 'model_name']}")
print(f"   For production: balance between {results.loc[best_test_idx, 'model_name']} (best accuracy)")
print(f"               and {results.loc[best_ratio_idx, 'model_name']} (best efficiency)")

print("\n" + "="*70)
```
